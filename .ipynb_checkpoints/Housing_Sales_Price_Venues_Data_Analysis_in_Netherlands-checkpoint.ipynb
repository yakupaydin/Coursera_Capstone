## Foursquare extract venues

I used foursquare venue explorer api to find venues near to city center. I choose 10000 meter as radius because citites has big araea to trace. At the same time, I extract 3 times with offset to get about 300 houndred venue per city.

```python
import requests
CLIENT_ID = 'HOFZJ1KBNJRXXBLO1XLXSZLBNPGVW11G5MO551W3W1O1FJSW' # your Foursquare ID
CLIENT_SECRET = 'RTPC5KJ2U5JWPLRF2RH1X5VP20FRJWDMRWZVQHZQTSQPZOR4' # your Foursquare Secret
VERSION = '20180604'
LIMIT = 100
RADIUS=10000
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=10000):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        offset=0
        for _ in range(3):   
            # create the API request URL
            url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&offset={}'.format(
                CLIENT_ID, 
                CLIENT_SECRET, 
                VERSION, 
                lat, 
                lng, 
                radius, 
                LIMIT,offset)
            offset+=100
            # make the GET request
            results = requests.get(url).json()["response"]['groups'][0]['items']
            
            venues_list.append([(
                    name, 
                    lat, 
                    lng, 
                    v['venue']['name'], 
                    v['venue']['location']['lat'], 
                    v['venue']['location']['lng'],  
                    v['venue']['categories'][0]['name']) for v in results])

        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Neighborhood', 
                      'Neighborhood Latitude', 
                      'Neighborhood Longitude', 
                      'Venue', 
                      'Venue Latitude', 
                      'Venue Longitude', 
                      'Venue Category']
    
    return(nearby_venues)
```

```python
netherlands_venues = getNearbyVenues(
    names=df['City'],
    latitudes=df['Latitude'],
    longitudes=df['Longitude']
)
```

# Methodology

**1**-As a database, I used GitHub repository in my study. My master data which has the main components Borough, Average House Price, Latitude and Longitude informations of the city.

**2**- I used python folium library to visualize geographic details of Istanbul and its boroughs and I created a map of Istanbul with boroughs superimposed on top. I used latitude and longitude values to get the visual

**3**- I utilized the Foursquare API to explore the boroughs and segment them. I designed the limit as 100 venue and the radius 10000 meter for each borough from their given latitude and longitude informations. 

```python
netherlands_onehot=pd.get_dummies(netherlands_venues[['Venue Category']],prefix="",prefix_sep="")

netherlands_onehot.insert(loc=0,column='City',value=netherlands_venues['Neighborhood'])
netherlands_onehot.head()
```

```python
netherlands_grouped=netherlands_onehot.groupby('City').mean().reset_index()
netherlands_grouped.head()
```

```python
netherlands_grouped.shape
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['City']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
netherlands_venues_sorted = pd.DataFrame(columns=columns)
netherlands_venues_sorted['City'] = netherlands_grouped['City']

for ind in np.arange(netherlands_grouped.shape[0]):
    netherlands_venues_sorted.iloc[ind, 1:] = return_most_common_venues(netherlands_grouped.iloc[ind, :], num_top_venues)

netherlands_venues_sorted.head()
```

# Analysis

We have some common venue categories in boroughs. In this reason I used unsupervised learning K-means algorithm to cluster the boroughs. K-Means algorithm is one of the most common cluster method of unsupervised learning.

First, I will run K-Means to cluster the boroughs into 3 clusters because when I analyze the K-Means with elbow method it ensured me the 7 degree for optimum k of the K-Means.

```python
from sklearn.cluster import KMeans
```

```python
df_clustering=netherlands_grouped.drop('City',1)
```

```python
Sum_of_squared_distances = []
K = range(1,10,1)
for k in K:
    km = KMeans(n_clusters=k,random_state=0)
    km = km.fit(df_clustering)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
k=6

km=KMeans(n_clusters=k, random_state=0).fit(df_clustering)

km.labels_
```

```python
df['Cluster']=km.labels_

df_merged=df

df_merged = df_merged.join(netherlands_venues_sorted.set_index('City'), on='City')

df_merged
```

**You can now see Join, Labels and Level_labels columns as the last three ones in above table. You can also see a clustered map boroughs of Istanbul in the below.**
